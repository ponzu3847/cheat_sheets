```
import statistics
```

```
import numpy as np
```

```
'''
statistics
.mean()

  機能：
    与えられた配列の平均値を返す
  使い方：
    statistics.mean(''配列)
'''
a=np.arange(10)
print(a)
print(statistics.mean(a))
```

```
'''
statistics
.median()

  機能：
    与えられた配列の中央値を返す
  使い方：
    statistics.median('配列')
'''
a=np.arange(10)
print(a)
print(statistics.median(a))
```

```
'''
statistics
.mode()

  機能：
    与えられた配列の最頻値を返す
  使い方：
    statistics.mode('配列')
'''
a=np.random.randint(0,10,10)
print(a)
print(statistics.mode(a))
```
