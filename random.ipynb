```python
import random
```

```python
'''
random
.randint()

  機能：
    与えられた２つの整数の範囲から一つの整数をランダムに発生させる
  使い方：
    random.randint(0,100)
'''
print(random.randint(0,100))
```

```python
'''
random
.shuffle()

    機能
        与えられた配列の要素をシャッフルする
    使い方
        random.shuffle(配列)
'''
a=[1,2,3,4,5]
random.shuffle(a)
print(a)
```

```python
'''
random
.sample()

    機能
        与えられた配列の要素をシャッフルして新しい配列を生成する
    使い方
        random.sample(配列,取得したい要素の個数)
'''
a=[1,2,3,4,5]
b=random.sample(a,len(a))
c=random.sample(a,2)
print('a=',a)
print('b=',b)
print('c=',c)
```
