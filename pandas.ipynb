### pandas.plotting

#### .scatter_matrix
(frame, alpha=0.5, figsize=None, ax=None, grid=False, diagonal='hist', marker='.', density_kwds=None, hist_kwds=None, range_padding=0.05, **kwargs)  
https://pandas.pydata.org/docs/reference/api/pandas.plotting.scatter_matrix.html?highlight=plotting%20scatter

```python
from sklearn.datasets import load_iris

iris_dataset=load_iris()
iris_dataFrame=pd.DataFrame(iris_dataset.data,columns=iris_dataset.feature_names)
grr=pd.plotting.scatter_matrix(iris_dataFrame,c=iris_dataset.target,marker="o",figsize=(15,15),range_padding=0.05,cmap='jet')
```
