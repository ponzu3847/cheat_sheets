```python
#f(x1,x2)=x1^2+x2^2
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3#3D

def func(x1,x2):
    '''
    3
    '''
    return x1**2+x2**2

######
if __name__=='__main__':
    x1=np.arange(-3,3,0.25)
    x2=np.arange(-3,3,0.25)
    X1,X2=np.meshgrid(x1,x2)#2
    Y=func(X1,X2)#y
    
    fig=plt.figure()#2
    ax=p3.Axes3D(fig)#3
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_zlabel('$f(x_1,x_2)$')
    ax.plot_wireframe(X1,X2,Y)
    plt.show()
```
