```python
#f(x1,x2)=x1^2+x2^2のグラフを描画する
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3#3Dモジュールのインポート

def func(x1,x2):
    '''
    3次元グラフで使用する関数
    '''
    return x1**2+x2**2

###実行ブロック###
if __name__=='__main__':
    x1=np.arange(-3,3,0.25)
    x2=np.arange(-3,3,0.25)
    X1,X2=np.meshgrid(x1,x2)#2次元の格子を生成
    Y=func(X1,X2)#格子のy値を取得
    
    fig=plt.figure()#2次元のグラフを生成
    ax=p3.Axes3D(fig)#3次元化
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_zlabel('$f(x_1,x_2)$')
    ax.plot_wireframe(X1,X2,Y)
    plt.show()
```
