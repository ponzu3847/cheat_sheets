#### .subplot
matplotlib.pyplot.subplot(*args, **kwargs)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html?highlight=subplot#matplotlib.pyplot.subplot  

現在の描画キャンパスにサブプロットを追加する  


```python
import matplotlib.pyplot as plt
import numpy as np


def linear(x):
    return x

def d_linear(x):
    return np.ones(*x.shape)

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return np.where(x > 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def d_tanh(x):
    return 1 - np.tanh(x) ** 2

x = np.arange(-5, 5, 0.01)

plt.figure(figsize=(10, 10),facecolor='w')

plt.subplot('221')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, linear(x))
plt.plot(x, d_linear(x))
plt.title('linear')
plt.grid(True)

plt.subplot('222')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, relu(x))
plt.plot(x, d_relu(x))
plt.title('relu')
plt.grid(True)

plt.subplot('223')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, sigmoid(x))
plt.plot(x, d_sigmoid(x))
plt.title('sigmoid')
plt.grid(True)

plt.subplot('224')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, tanh(x))
plt.plot(x, d_tanh(x))
plt.title('tanh')
plt.grid(True)

plt.show()
```

#### .subplots
matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html?highlight=subplots#matplotlib.pyplot.subplots  


```python
import matplotlib.pyplot as plt
import numpy as np


def linear(x):
    return x

def d_linear(x):
    return np.ones(*x.shape)

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return np.where(x > 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def d_tanh(x):
    return 1 - np.tanh(x) ** 2

x = np.arange(-5, 5, 0.01)
func_list=[(linear,d_linear),(relu,d_relu),(sigmoid,d_sigmoid),(tanh,d_tanh)]

fig,axes=plt.subplots(2,2,figsize=(10, 10),facecolor='w')

for ax,func in zip(axes.ravel(),func_list):
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.plot(x, func[0](x))
    ax.plot(x, func[1](x))
    ax.set_title(func[0].__name__)
    ax.grid(True)

plt.show()
```

#### .colorbar()
機能：
* カラーバーを表示する
  
使い方例：
* c=ax.contour(x,y,z)
* fig.colorbar(c)

```python
import numpy as np
import matplotlib.pyplot as plt


x=np.linspace(-1,1,5)
y=np.linspace(-1,1,5)
X,Y=np.meshgrid(x,y)
Z=np.sqrt(X**2+Y**2)

fig=plt.figure(figsize=(24,8),facecolor='w')
ax=fig.add_subplot('131',projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X,Y,Z,cmap='plasma')

ax=fig.add_subplot('132',projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X,Y,Z,cmap='viridis')

ax=fig.add_subplot('133')
ax.set_xlabel('x')
ax.set_ylabel('y')
c=ax.contour(X,Y,Z,levels=5)
fig.colorbar(c)
c.clabel(fmt='%1.1f',fontsize=14)
plt.show()
```

#### .quiver
matplotlib.pyplot.quiver(*args, data=None, **kwargs)  
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html  

* グラフに二次元のベクトルを追加する

```python
a=np.random.randint(0,10,11)
b=np.random.randint(0,10,11)
aa=np.array(a)
bb=np.array(b)
print(f'aa={aa}')
print(f'bb={bb}')
print(f'aa[1:]={aa[1:]}')
print(f'aa[:-1]={aa[:-1]}')
print(f'aa[1:]-aa[:-1]={aa[1:]-aa[:-1]}')
fig=plt.figure(figsize=(6,6),facecolor='w')
ax=fig.add_subplot('111')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_xlim(0,10)
ax.set_ylim(0,10)
C=(aa[1:]-aa[:-1])**2+(bb[1:]-bb[:-1])**2   #ベクトルの大きさを示す値（ベクトルカラー表示用）
ax.quiver(aa[:-1],bb[:-1],aa[1:]-aa[:-1],bb[1:]-bb[:-1],C,scale_units='xy',angles='xy',scale=1)
ax.grid(True)
```

#### .hist
matplotlib.pyplot.hist(x, bins=10, range=None, normed=False, weights=None,
                       cumulative=False, bottom=None, histtype='bar',
                       align='mid', orientation='vertical', rwidth=None,
                       log=False, color=None, label=None, stacked=False,
                       hold=None, data=None, **kwargs)  
https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.hist.html  
  
ヒストグラムを作成する
* x (必須):	ヒストグラムを作成するための生データの配列。
* bins:	ビン (表示する棒) の数。階級数。(デフォルト値: 10)
* range:	ビンの最小値と最大値を指定。(デフォルト値: (x.min(), x.max()))
* normed:	True に設定すると正規化 (合計値が 1 になるように変換) を実施。 (デフォルト値: False)
* cumulative:	True に設定すると、累積ヒストグラムを出力。 (デフォルト値: False)
* bottom:	各棒の下側の余白を数値または配列で指定。
* histtype:	‘bar’ (通常のヒストグラム), ‘barstacked’ (積み上げヒストグラム), ‘step’ (線), ‘stepfilled ‘ (塗りつぶしありの線) から選択。 (デフォルト値: ‘bar’)
* align:	各棒の中心を X 軸目盛上のどの横位置で出力するか。 ‘left’, ‘mid’, ‘right’ から選択。(デフォルト値: ‘mid’)
* orientation:	棒の方向。’horizontal’ (水平方向), ‘vertical’ (垂直方向) から選択。(デフォルト値: ‘vertical’)
* rwidth:	各棒の幅を数値または、配列で指定。
* log:	True に設定すると、縦軸を対数目盛で表示します。
* color:	ヒストグラムの色。配列で指定し、データセット単位で色を指定することができます。
* label:	凡例を載せる際に使用します。
* stacked:	True に設定すると積み上げヒストグラムで出力します。False に設定すると、横に並べて出力します。

```python
x=np.random.randn(1000)
plt.hist(x,bins=50)
plt.show()
```

#### .annotate
matplotlib.pyplot.annotate(text, xy, *args, **kwargs)  
https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.annotate.html  
  
ポイントxyにテキストで注釈を付けます。  


```python
x=np.random.randn(10)
y=np.random.randn(10)
text=['a'+str(i) for i in range(len(x))]
print('x:',x)
print('y:',y)
print('text:',text)

plt.xlim(-2,2)
plt.ylim(-2,2)
for i in range(len(x)):
    plt.annotate(text[i],(x[i],y[i]))
plt.scatter(x,y)
plt.show()
```

#### .text
matplotlib.pyplot.text(x, y, s, fontdict=None, **kwargs)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html?highlight=text#matplotlib.pyplot.text  


```python
import numpy as np
import matplotlib.pyplot as plt

n_samples=10
x=np.array(np.random.randint(0,10,size=(n_samples,2)))
y=np.arange(n_samples)

print(x)

plt.xlim(x[:,0].min()-1,x[:,0].max()+1)
plt.ylim(x[:,1].min()-1,x[:,1].max()+1)
for i in range(n_samples):
    plt.text(x[i,0],x[i,1],str(y[i]),c='r')
```
