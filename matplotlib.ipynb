#### .subplot
matplotlib.pyplot.subplot(*args, **kwargs)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html?highlight=subplot#matplotlib.pyplot.subplot  

現在の描画キャンパスにサブプロットを追加する  


```python
import matplotlib.pyplot as plt
import numpy as np


def linear(x):
    return x

def d_linear(x):
    return np.ones(*x.shape)

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return np.where(x > 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def d_tanh(x):
    return 1 - np.tanh(x) ** 2

x = np.arange(-5, 5, 0.01)

plt.figure(figsize=(10, 10),facecolor='w')

plt.subplot('221')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, linear(x))
plt.plot(x, d_linear(x))
plt.title('linear')
plt.grid(True)

plt.subplot('222')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, relu(x))
plt.plot(x, d_relu(x))
plt.title('relu')
plt.grid(True)

plt.subplot('223')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, sigmoid(x))
plt.plot(x, d_sigmoid(x))
plt.title('sigmoid')
plt.grid(True)

plt.subplot('224')
plt.xlim(-2,2)
plt.ylim(-2, 2)
plt.plot(x, tanh(x))
plt.plot(x, d_tanh(x))
plt.title('tanh')
plt.grid(True)

plt.show()
```

#### .subplots
matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html?highlight=subplots#matplotlib.pyplot.subplots  


```python
import matplotlib.pyplot as plt
import numpy as np


def linear(x):
    return x

def d_linear(x):
    return np.ones(*x.shape)

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return np.where(x > 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def d_tanh(x):
    return 1 - np.tanh(x) ** 2

x = np.arange(-5, 5, 0.01)
func_list=[(linear,d_linear),(relu,d_relu),(sigmoid,d_sigmoid),(tanh,d_tanh)]

fig,axes=plt.subplots(2,2,figsize=(10, 10),facecolor='w')

for ax,func in zip(axes.ravel(),func_list):
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.plot(x, func[0](x))
    ax.plot(x, func[1](x))
    ax.set_title(func[0].__name__)
    ax.grid(True)

plt.show()
```

```python
'''
matplotlib.pyplot
.figure()

  機能：
    新しい図を作成するか、既存の図をアクティブにする
  使い方：
    plt.figure(figsize=('幅','高さ'),
              facecolor='背景色',
              edgecolor='ボーダー色',
              dpi='インチあたりのドット数（解像度）')

.add_subplot()

    機能：
      指定した描画キャンバスにグラフを追加してアクティブにする
    使い方例：
      fig=plt.figure(figsize=(3,3))
      ax=fig.add_subplot('121',   #１行２列のグラフのうち１番目に３次元グラフを追加する
                          projection='3d')   #３次元グラフ
'''
def f(x):
  return 2*x

def g(x):
  return x**2

x0=np.linspace(0,20,100)

fig=plt.figure(figsize=(12,3),facecolor='w',edgecolor='k',dpi=100)
ax=fig.add_subplot('121')
ax.plot(x0,f(x0),linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('f(x)=2x')
ax.grid(True)

ax=fig.add_subplot('122')
ax.plot(x0,g(x0),linestyle='solid')
ax.set_xlabel('x')
ax.set_ylabel('g(x)')
ax.set_title('$g(x)=x{^%d}$'%2)

plt.show()
```

```python
'''
matplotlib.pyplot
.plot_surface()

  機能：
    ３次元グラフに面を表示する
  使い方例：
    fig=plt.figure(figsize=())
    ax=fig.add_subplot('111',projection='3d')
    ax.plot_surface(x,y,z,cmap='カラーマップの種類')

.contour(),.contourf()
  機能：
    グラフに等高線を表示する(contoufは塗りつぶし)
  使い方：
    c=ax.contour(x,y,z,levels=等高線の間隔,cmap='カラーマップの種類')

.clabel()
  機能：
    グラフ内の等高線に値を表示する
  使い方例：
    c=ax.contour(x,y,z)
    c.clabel(fmt='%1.1f',fontsize=14)

.colorbar()
  機能：
    カラーバーを表示する
  使い方例：
    c=ax.contour(x,y,z)
    fig.colorbar(c)
'''
x=np.linspace(-1,1,5)
y=np.linspace(-1,1,5)
X,Y=np.meshgrid(x,y)
Z=np.sqrt(X**2+Y**2)

fig=plt.figure(figsize=(24,8),facecolor='w')
ax=fig.add_subplot('131',projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X,Y,Z,cmap='plasma')

ax=fig.add_subplot('132',projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X,Y,Z,cmap='viridis')

ax=fig.add_subplot('133')
ax.set_xlabel('x')
ax.set_ylabel('y')
c=ax.contour(X,Y,Z,levels=5)
fig.colorbar(c)
c.clabel(fmt='%1.1f',fontsize=14)
plt.show()
```

```python
'''
matplotlib.pyplot
.quiver()
  https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html

  機能：
    グラフに二次元のベクトルを追加する
  使い方：
    plt.quiver([X,Y],U,V,[C],**kw)
      X,Y:  ベクトルの開始点のxy座標(配列または数値)
      U,V:  ベクトルの成分(配列または数値)
      C:  ベクトルのカラー（大きさ）を表すための配列
'''
a=np.random.randint(0,10,11)
b=np.random.randint(0,10,11)
aa=np.array(a)
bb=np.array(b)
print(f'aa={aa}')
print(f'bb={bb}')
print(f'aa[1:]={aa[1:]}')
print(f'aa[:-1]={aa[:-1]}')
print(f'aa[1:]-aa[:-1]={aa[1:]-aa[:-1]}')
fig=plt.figure(figsize=(6,6),facecolor='w')
ax=fig.add_subplot('111')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_xlim(0,10)
ax.set_ylim(0,10)
C=(aa[1:]-aa[:-1])**2+(bb[1:]-bb[:-1])**2   #ベクトルの大きさを示す値（ベクトルカラー表示用）
ax.quiver(aa[:-1],bb[:-1],aa[1:]-aa[:-1],bb[1:]-bb[:-1],C,scale_units='xy',angles='xy',scale=1)
ax.grid(True)
```

```python
'''
matplotlib.pyplot
.annotate()
https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.annotate.html

機能：
    ポイントxyにテキストで注釈を付けます。
    最も単純な形式では、テキストはxyに配置されます。
    オプションで、テキストを別の位置tに表示できます。
    次に、arrowpropsを定義することにより、テキストから注釈付きの点xyを指す矢印を追加できます。
使い方：
    matplotlib.pyplot.annotate(text, xy, *args, **kwargs)
'''
x=np.random.randn(10)
y=np.random.randn(10)
text=['a'+str(i) for i in range(len(x))]
print('x:',x)
print('y:',y)
print('text:',text)

plt.xlim(-2,2)
plt.ylim(-2,2)
for i in range(len(x)):
    plt.annotate(text[i],(x[i],y[i]))
plt.scatter(x,y)
plt.show()
```
