numpy.identity(n, dtype=None, *, like=None)  
https://numpy.org/doc/stable/reference/generated/numpy.identity.html?highlight=ident  

* 単位行列を作成する

```python
#単位行列とゼロ行列と転置行列と逆行列
a=np.array([[2,3,4],
         [5,6,7],
         [8,9,1]])

inv=np.linalg.inv(a)

zeros=np.zeros((3,3))#3*3のゼロ行列を作成
zeros_like=np.zeros_like(a)
unit=np.identity(3)#3*3の単位行列を作成

print('a=\n',a)
print('zeros(3,3)=\n',zeros)
print('zeros_like(a)=\n',zeros_like)
print('identity(3)=\n',unit)

print('a*zeros=\n',a*zeros)
print('np.dot(a,unit)=\n',np.dot(a,unit))

print('a.T=\n',a.T)#aの転置行列

print('a.inv=\n',inv)#aの逆行列
print('np.dot(a,inv)=\n',np.dot(a,inv))
```

### .r_
numpy.r_ = <numpy.lib.index_tricks.RClass object>  
https://numpy.org/doc/stable/reference/generated/numpy.r_.html?highlight=r_#numpy.r_

* 多次元配列を結合する（np.c_はnp.r_の特殊な場合を抜き出したもの）
* 特に何も指定しない場合は、np.r_はhstack、np.c_はvstackと同様の使い方をすることができる
* np.r_[配列１,配列２]
* np.r_[start:stop:step]
* np.r_['結合する軸（axis=0〜）の指定',配列１,配列２]

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6]) # まずは１次元配列で
print(np.r_[a,b])
print(np.r_[2, 5, 3, np.array([2,3,]), 4.2])
c = np.zeros((2,3)) # 2次元配列を結合する。axis=1方向の要素数は一致していないと結合できないことに注意
d = np.ones((3, 3)) # cのaxis=1方向の要素数は3なので、それに合わせる。axis=0方向の要素数はcと同じ2である必要はない
print(np.r_[c, d])

print(np.r_[1:10:1])
print(np.r_[10:1:-1])
print(np.r_[1:10:10j])#10等分
print(np.r_[1:5:1,np.array([3,3])])

e=np.ones((2,2))#2次元配列
f=np.zeros((2,2))
print(np.r_['0',e,f])#結合する方向を指定
print(np.r_['1',e,f])

g=np.ones((2,2,2))#３次元配列
print('g=\n',g)
h=np.zeros((2,2,2))
print('h=\n',h)
print('axis=0に結合->\n',np.r_['0',g,h])
print('shape->',np.r_['0',g,h].shape,'\n')
print('axis=1に結合->\n',np.r_['1',g,h])
print('shape->',np.r_['1',g,h].shape,'\n')
print('axis=2に結合->\n',np.r_['2',g,h])
print('shape->',np.r_['2',g,h].shape,'\n')
```

### .digitize
numpy.digitize(x, bins, right=False)  
https://numpy.org/doc/stable/reference/generated/numpy.digitize.html?highlight=digitize#numpy.digitize

どのビンにデータポイントが入っているかを返す

```python
import numpy as np
import mglearn
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import Ridge

X,y=mglearn.datasets.make_wave(n_samples=40)
bins=np.linspace(-3,3,11)
display(bins)

which_bin=np.digitize(X,bins=bins)
display(X[:5])
display(which_bin[:5])

encoder=OneHotEncoder(sparse=False)
X_binned=encoder.fit_transform(which_bin)

display(X_binned[:5])

ridge=Ridge()
ridge.fit(X_binned,y)

line=np.linspace(-3,3,1000,endpoint=False).reshape(-1,1)
line_binned=encoder.transform(np.digitize(line,bins=bins))

plt.scatter(X,y)
plt.vlines(bins,ymin=-3,ymax=3,linestyle='--')
plt.plot(line,ridge.predict(line_binned))
plt.show()
```

### .meshgrid

https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html?highlight=meshgrid#numpy.meshgrid

* 座標ベクトルから座標行列を返す
* numpy.meshgrid（座標ベクトル,copy = True,sparse = False,indexing = 'xyかij' ）
* copy:  Falseにすると、メモリを節約するために、元の配列のビューを返します。

### .nditer
class numpy.nditer(op, flags=None, op_flags=None, op_dtypes=None, order='K', casting='safe', op_axes=None, itershape=None, buffersize=0)  
https://numpy.org/doc/stable/reference/generated/numpy.nditer.html

* 配列のインデックスをタプル形式で順番に返すイテレータ

```python
#一次元配列の場合
x=np.array([1,2,3,4])
it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
print('x=',x)
while not it.finished:
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()

#二次元配列の場合
x=np.array([[1,2],[3,4]])
it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
print('x=\n',x)
while not it.finished:
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()

#op_flags=['readwrite']の場合
x=np.array([[1,2],[3,4]])
it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
print('x=\n',x)
while not it.finished:
  it[0]=99    #readwriteの場合は書き換え可
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()
print('x=\n',x)

#op_flags=['readonlly']の場合
x=np.array([[1,2],[3,4]])
it=np.nditer(x,flags=['multi_index'],op_flags=['readonly'])
print('x=\n',x)
while not it.finished:
  it[0]=99    #readonlyの場合は書き換え不可のためエラーになる
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()
print('x=\n',x)
```
