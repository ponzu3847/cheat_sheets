```python
'''
numpy
.arange()
    機能
        公差を指定して等差数列を作成する
    使い方
        np.arange(最初,最後,公差)
    例
        np.arange(1,1001,2)
        
.ones()
    機能
        成分が１のベクトルを作成する
    使い方
        a=np.ones(作成するベクトル成分の数)
        a=np.ones([行数,列数])
'''
#Σが示す和（１から１００までの和を求める）の計算を内積として計算
a=np.ones(100)#1を１００個並べた行ベクトルを作る
b=np.arange(1,101)#1から１００までの等差数列を作る
print(a)
print(np.dot(a,b))
```

```python
#単位行列とゼロ行列と転置行列と逆行列
'''
numpy
.T
    機能
        行列の転置行列を求める
    使い方
        a=np.array([[1,2],[3,4],[5,6]])
        a.T
        
.zeros()
    機能
        ゼロ行列を作成する
    使い方
        np.zeros((行数,列数))
    例
        np.zeros((3,3))

.zeros_like()

    機能：
        対象の配列とおなじ形状のゼロ行列を作成する
    使い方：
        np.zeros_like(配列)

.identity()
    機能
        単位行列を作成する
    使い方
        np.identity(行数)#単位行列は行数＝列数
    例
        np.identity(3)
        
.linalg.inv()
    機能
        行列の逆行列を求める
    使い方
        a=np.array([[1,2],[3,4],[5,6]])
        np.linalg.inv(a)
'''
a=np.array([[2,3,4],
         [5,6,7],
         [8,9,1]])

inv=np.linalg.inv(a)

zeros=np.zeros((3,3))#3*3のゼロ行列を作成
zeros_like=np.zeros_like(a)
unit=np.identity(3)#3*3の単位行列を作成

print('a=\n',a)
print('zeros(3,3)=\n',zeros)
print('zeros_like(a)=\n',zeros_like)
print('identity(3)=\n',unit)

print('a*zeros=\n',a*zeros)
print('np.dot(a,unit)=\n',np.dot(a,unit))

print('a.T=\n',a.T)#aの転置行列

print('a.inv=\n',inv)#aの逆行列
print('np.dot(a,inv)=\n',np.dot(a,inv))
```

```python
'''
numpy
.r_[]
  https://deepage.net/features/numpy-cr.html

  機能：
    多次元配列を結合する（np.c_はnp.r_の特殊な場合を抜き出したもの）
    特に何も指定しない場合は、np.r_はhstack、np.c_はvstackと同様の使い方をすることができる
  使い方：
    np.r_[配列１,配列２]
    np.r_[start:stop:step]
    np.r_['結合する軸（axis=0〜）の指定',配列１,配列２]
'''
a = np.array([1, 2, 3])
b = np.array([4, 5, 6]) # まずは１次元配列で
print(np.r_[a,b])
print(np.r_[2, 5, 3, np.array([2,3,]), 4.2])
c = np.zeros((2,3)) # 2次元配列を結合する。axis=1方向の要素数は一致していないと結合できないことに注意
d = np.ones((3, 3)) # cのaxis=1方向の要素数は3なので、それに合わせる。axis=0方向の要素数はcと同じ2である必要はない
print(np.r_[c, d])

print(np.r_[1:10:1])
print(np.r_[10:1:-1])
print(np.r_[1:10:10j])#10等分
print(np.r_[1:5:1,np.array([3,3])])

e=np.ones((2,2))#2次元配列
f=np.zeros((2,2))
print(np.r_['0',e,f])#結合する方向を指定
print(np.r_['1',e,f])

g=np.ones((2,2,2))#３次元配列
print('g=\n',g)
h=np.zeros((2,2,2))
print('h=\n',h)
print('axis=0に結合->\n',np.r_['0',g,h])
print('shape->',np.r_['0',g,h].shape,'\n')
print('axis=1に結合->\n',np.r_['1',g,h])
print('shape->',np.r_['1',g,h].shape,'\n')
print('axis=2に結合->\n',np.r_['2',g,h])
print('shape->',np.r_['2',g,h].shape,'\n')
```

```python
'''
nunmpy.random
.choice()
https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html

機能：
    指定された1次元配列からランダムサンプルを生成します
使い方：
    numpy.random.choice(a, size=None, replace=True, p=None)
    numpy.random.choice(配列,size=サイズ,replace=重複有無,p=確率分布配列)
'''
def softmax(x):
    out=np.exp(x)/np.sum(np.exp(x))
    return out

x=np.random.randint(5,size=5)
print('x\n',x)
p=softmax(x)
print('\np\n',p)
y=['a','b','c','d','e']
print('\ny\n',y)
z=np.random.choice(y,size=5,p=p)
print('\nz\n',z)
```

```python
'''
numpy
.meshgrid()
  https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html
  https://deepage.net/features/numpy-meshgrid.html

  機能：
    座標ベクトルから座標行列を返す
  使い方：
    numpy.meshgrid（座標ベクトル,copy = True,sparse = False,indexing = 'xyかij' ）
    ※copy:  Falseにすると、メモリを節約するために、元の配列のビューを返します。
    ※sparce:  デフォルトはFalse,Trueの場合、メモリを節約するためにスパースグリッドの配列を返す
    ※indexing:  返す配列の形状を指定する（ｘｙ軸順か行列順か）
'''
x = np.arange(-2,2,0.1)
y = np.arange(-2,2,0.1)
xx, yy = np.meshgrid(x, y) # 格子点となるx,y座標を作成。  
print('xx=\n',xx)
print('yy=\n',yy)
ret = yy-xx**3
print('ret=\n',ret)
def plot1():
    plt.gca().set_aspect('equal', adjustable='box') # グラフの縦横の比をそろえるコマンド   
    plt.contourf(x, y, ret>=0, cmap=plt.cm.bone) # 条件(ret>=0)を満たす部分を白、満たさない部分を黒とする。
    plt.show()
plot1()
```

```python
'''
numpy
.nditer()
https://snowtree-injune.com/2020/06/29/nditer-z009/
https://numpy.org/doc/stable/reference/generated/numpy.nditer.html

機能：
  配列のインデックスをタプル形式で順番に返すイテレータ
使い方例：
  np.nditer(配列,flags=['multi_index'],op_flags=['readwrite])
'''
#一次元配列の場合
x=np.array([1,2,3,4])
it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
print('x=',x)
while not it.finished:
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()

#二次元配列の場合
x=np.array([[1,2],[3,4]])
it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
print('x=\n',x)
while not it.finished:
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()

#op_flags=['readwrite']の場合
x=np.array([[1,2],[3,4]])
it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
print('x=\n',x)
while not it.finished:
  it[0]=99    #readwriteの場合は書き換え可
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()
print('x=\n',x)

#op_flags=['readonlly']の場合
x=np.array([[1,2],[3,4]])
it=np.nditer(x,flags=['multi_index'],op_flags=['readonly'])
print('x=\n',x)
while not it.finished:
  it[0]=99    #readonlyの場合は書き換え不可のためエラーになる
  idx=it.multi_index
  print('idx=',idx,'x[idx]=',x[idx],'it[0]=',it[0])
  it.iternext()
print('x=\n',x)
```

```python
'''
numpy
.sort(),.argsort()
https://numpy.org/doc/stable/reference/generated/numpy.sort.html
https://numpy.org/doc/stable/reference/generated/numpy.argsort.html
https://note.nkmk.me/python-numpy-sort-argsort/

.sort()
機能：
    ソートされた新たなndarrayが返され、元のndarrayはそのまま。
    一次元配列の場合は単純に昇順にソートされる。
使い方：
    numpy.sort(a, axis=-1, kind=None, order=None)

.argsort()
機能：
    np.argsort()は値ではなく並び替えたインデックス（元のndarrayでの位置 = 0始まりの順番）のndarrayを返す。
    引数axisなどの考え方はnp.sort()と同じ。
使い方：
    numpy.argsort(a, axis=-1, kind=None, order=None)
'''
x=np.array([5,3,6,8,2,4,6])
y=np.sort(x)
z=np.argsort(x)

print('x=',x)
print('y=',y)
print('z=',z)

sort_list=[]
for i in z:
    sort_list.append(x[i])

print(sort_list)
```

### .digitize
numpy.digitize(x, bins, right=False)  
https://numpy.org/doc/stable/reference/generated/numpy.digitize.html?highlight=digitize#numpy.digitize

どのビンにデータポイントが入っているかを返す

```python
import numpy as np
import mglearn
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import Ridge

X,y=mglearn.datasets.make_wave(n_samples=40)
bins=np.linspace(-3,3,11)
display(bins)

which_bin=np.digitize(X,bins=bins)
display(X[:5])
display(which_bin[:5])

encoder=OneHotEncoder(sparse=False)
X_binned=encoder.fit_transform(which_bin)

display(X_binned[:5])

ridge=Ridge()
ridge.fit(X_binned,y)

line=np.linspace(-3,3,1000,endpoint=False).reshape(-1,1)
line_binned=encoder.transform(np.digitize(line,bins=bins))

plt.scatter(X,y)
plt.vlines(bins,ymin=-3,ymax=3,linestyle='--')
plt.plot(line,ridge.predict(line_binned))
plt.show()
```
