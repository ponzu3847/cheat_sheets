##pickle
https://docs.python.org/ja/3/library/pickle.html

```
'''
pickle
.dump()

機能：
  オブジェクト（配列、変数など）をひとつのまとまり（ファイル）として保存する
使い方：
  pickle.dump(オブジェクト,ファイル)
'''
a=[111,'abc',{'k1':34}]
with open('/content/drive/My Drive/Colab Notebooks/cheatsheets/data/test.pkl','wb') as f:
  pickle.dump(a,f)
```

```
'''
pickle
.load()

機能：
  pickleで保存したデータを読み込む
使い方：
  pickle.load('pklファイル')
'''
with open('/content/drive/My Drive/Colab Notebooks/cheatsheets/data/test.pkl','rb') as f:
  a=pickle.load(f)
  print(a)
  print(a[2]['k1'])
```
