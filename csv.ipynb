```
import csv
```

```
'''
csv
  https://docs.python.org/ja/3/library/csv.html

.reader()
  機能：
    与えられたcsvfile内の行を反復処理するようなreaderオブジェクトを返す
    イテレータ(iterator)プロトコルに対応したオブジェクトなのでfor文やnextで行データを取得することが可能
    行データはリストなのでインデックスでアクセスできる
  使い方：
    data=csv.reader('読み込むファイルの場所',delimiter='区切り文字',quotechar='フィールドを囲むための文字')
'''
with open('/content/drive/My Drive/Colab Notebooks/cheatsheets/data/access.csv', newline='') as csvfile:
  reader = csv.reader(csvfile,delimiter=',',quotechar='"')

  for row in reader:
    print(row[0])
```

```
'''
csv
  https://docs.python.org/ja/3/library/csv.html

.DictReader()
  機能：
    イテレータ(iterator)プロトコルに対応したオブジェクトで返却される
    行データが辞書型で表現されているため、カラム名を指定して取得することができる
    カラム名は自動的に一行目が読み込まれる
  使い方：
    data=csv.DictReader('ファイルの場所')
'''
with open('/content/drive/My Drive/Colab Notebooks/cheatsheets/data/access.csv', newline='') as csvfile:
  reader = csv.DictReader(csvfile)
  for row in reader:
    print(row)
```
