* é€šå¸¸æœ€å°äºŒä¹—æ³•ï¼ˆOLS: ordinary least squaresï¼‰ï¼š  
è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã«ãŠã„ã¦ã€äºˆæ¸¬ã¨çœŸã®å›žå¸°ã‚¿ãƒ¼ã‚²ãƒƒãƒˆyã¨ã®å¹³å‡äºŒä¹—èª¤å·®ï¼ˆmean squared errorï¼‰ãŒæœ€å°ã«ãªã‚‹ã‚ˆã†ã«ã€ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿wï¼ˆé‡ã¿ã€ã‚‚ã—ãã¯ä¿‚æ•°ï¼ˆcoefficientï¼‰ï¼‰ã¨bï¼ˆã‚ªãƒ•ã‚»ãƒƒãƒˆã€ã‚‚ã—ãã¯åˆ‡ç‰‡ï¼ˆintercept,bï¼‰ï¼‰ã‚’æ±‚ã‚ã‚‹


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import mglearn
from icecream import ic

X,y=mglearn.datasets.make_wave(n_samples=40)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
ic(X_train.shape)
ic(X_test.shape)
ic(y_train.shape)
ic(y_test.shape)

lr=LinearRegression().fit(X_train,y_train)
ic(lr.score(X_train,y_train))   # R^2(æ±ºå®šä¿‚æ•°)ã‚’è¡¨ç¤º
ic(lr.score(X_test,y_test)) # R^2(æ±ºå®šä¿‚æ•°)ã‚’è¡¨ç¤º
ic(lr.coef_)    # é‡ã¿
ic(lr.intercept_)   # åˆ‡ç‰‡
```

```python
import matplotlib.pyplot as plt
import numpy as np

line=np.linspace(-4,4,100)

plt.xlim(-3,3)
plt.scatter(X_train,y_train,color='b',label='train data',marker='o')
plt.scatter(X_test,y_test,color='r',label='test data',marker='^')
plt.plot(line,lr.coef_*line+lr.intercept_,color='black',label='predict',linestyle='--')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from icecream import ic

cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)
tree=DecisionTreeClassifier(random_state=0,max_depth=4)
tree.fit(X_train,y_train)
ic(tree.score(X_train,y_train))
ic(tree.score(X_test,y_test))
display(cancer.feature_names)
display(tree.feature_importances_)  # ç‰¹å¾´é‡ã®é‡è¦åº¦ã‚’è¡¨ç¤º
```

* ã™ã¹ã¦ã®æ±ºå®šæœ¨ã«ã‚ˆã‚‹å›žå¸°ãƒ¢ãƒ‡ãƒ«ã¯å¤–æŒ¿ï¼ˆextrapolateï¼‰ãŒã§ããªã„

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import mglearn
import matplotlib.pyplot as plt
from icecream import ic
import numpy as np

X,y=mglearn.datasets.make_wave(n_samples=100)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

tree=DecisionTreeRegressor(random_state=0,max_depth=2).fit(X_train,y_train)
ic(tree.score(X_train,y_train))
ic(tree.score(X_test,y_test))


line=np.linspace(-3,3,1000).reshape(-1,1)
plt.scatter(X_train,y_train,label='train_data',marker='o',color='b')
plt.scatter(X_test,y_test,label='test_data',marker='x',color='r')
plt.plot(line,tree.predict(line),linestyle='-',color='g',label='predict')
plt.legend()
plt.show()
```

é•·æ‰€ã€çŸ­æ‰€ã€ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿
* å›žå¸°ã§ã‚‚ã‚¯ãƒ©ã‚¹åˆ†é¡žã§ã‚‚ã€ãƒ©ãƒ³ãƒ€ãƒ ãƒ•ã‚©ãƒ¬ã‚¹ãƒˆãŒæœ€ã‚‚åºƒãä½¿ã‚ã‚Œã¦ã„ã‚‹æ©Ÿæ¢°å­¦ç¿’æ‰‹æ³•
* å¤šãã®å ´åˆã€ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ãƒãƒ¥ãƒ¼ãƒ‹ãƒ³ã‚°ã‚’ã›ãšã«ä½¿ãˆã€ãƒ‡ãƒ¼ã‚¿ã®ã‚¹ã‚±ãƒ¼ãƒ«å¤‰æ›ã‚‚å¿…è¦ãªã„

* å¤§ãã„ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã«å¯¾ã—ã¦ãƒ©ãƒ³ãƒ€ãƒ ãƒ•ã‚©ãƒ¬ã‚¹ãƒˆã‚’ä½œæˆã™ã‚‹ãŸã‚ã«ã¯æ™‚é–“ãŒã‹ã‹ã‚‹ãŒã€è¨ˆç®—æ©Ÿä¸Šã®CPUã‚³ã‚¢ï¼ˆn_jobsãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã§ã‚³ã‚¢æ•°ã‚’æŒ‡å®šï¼‰ã‚’è¤‡æ•°ä½¿ã£ã¦ä¸¦åˆ—åŒ–ã§ãã‚‹
* n_jobs=-1ã¨ã™ã‚‹ã¨è¨ˆç®—æ©Ÿä¸Šã®ã™ã¹ã¦ã®ã‚³ã‚¢ã‚’ä½¿ç”¨ã™ã‚‹
* åŸºæœ¬çš„ã«ãƒ©ãƒ³ãƒ€ãƒ æŽ¢ç´¢ã®ãŸã‚ã€random_stateã‚’å›ºå®šã—ãªã„ã¨æ§‹ç¯‰ã•ã‚Œã‚‹ãƒ¢ãƒ‡ãƒ«ãŒå¤§ããå¤‰ã‚ã‚‹ã“ã¨ã«æ³¨æ„
* ãƒ©ãƒ³ãƒ€ãƒ ãƒ•ã‚©ãƒ¬ã‚¹ãƒˆã¯ãƒ†ã‚­ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ãªã©ã®éžå¸¸ã«é«˜æ¬¡å…ƒã§ç–Žãªãƒ‡ãƒ¼ã‚¿ã«å¯¾ã—ã¦ã¯ã†ã¾ãæ©Ÿèƒ½ã—ãªã„å‚¾å‘ã«ã‚ã‚‹
* ç·šå½¢ãƒ¢ãƒ‡ãƒ«ã‚ˆã‚Šã‚‚å¤šãã®ãƒ¡ãƒ¢ãƒªã‚’æ¶ˆè²»ã—ã€è¨“ç·´ã‚‚äºˆæ¸¬ã‚‚é…ã„ãŸã‚ã€å®Ÿè¡Œæ™‚é–“ã‚„ãƒ¡ãƒ¢ãƒªãŒé‡è¦ãªã‚¢ãƒ—ãƒªã§ã¯ç·šå½¢ãƒ¢ãƒ‡ãƒ«ã‚’ä½¿ã£ãŸã»ã†ãŒè‰¯ã„
* n_estimatorsã¯å¤§ãã‘ã‚Œã°å¤§ãã„ã»ã©ã€ã‚ˆã‚Šå¤šãã®æ±ºå®šæœ¨ã®å¹³å‡ã‚’ã¨ã‚Šã€éŽå‰°é©åˆãŒä½Žæ¸›ã•ã‚Œã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«ãŒé ‘å¥ã«ãªã‚‹ã‚‹ãŸã‚ã€ã‚ˆã„ï¼ˆæ™‚é–“ã¨ãƒ¡ãƒ¢ãƒªã®ã‚ã‚‹é™ã‚Šå¤§ããã™ã‚‹ï¼‰
* max_featuresã¯å€‹ã€…ã®æ±ºå®šæœ¨ã®ä¹±æ•°æ€§ã‚’æ±ºå®šã™ã‚‹ã¨ã¨ã‚‚ã«ã€max_featuresãŒå°ã•ããªã‚‹ã¨éŽå‰°é©åˆãŒä½Žæ¸›ã™ã‚‹ï¼ˆä¸€èˆ¬ã«ã¯ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆå€¤ã‚’ä½¿ã†ã¨è‰¯ã„ï¼‰
* ã‚¯ãƒ©ã‚¹åˆ†é¡žã«ã¤ã„ã¦ã¯ã€max_features=sqrt(n_features)ã€å›žå¸°ã«ã¤ã„ã¦ã¯max_features=n_featuresã¨ãªã£ã¦ã„ã‚‹
* max_featuresã‚„max_leaf_nodesã‚’è¿½åŠ ã™ã‚‹ã¨æ€§èƒ½ãŒä¸ŠãŒã£ãŸã‚Šã€è¨“ç·´ã‚„äºˆæ¸¬ã«ã‹ã‹ã‚‹æ™‚é–“ãŒå¤§å¹…ã«ç¸®ã¾ã‚‹ã“ã¨ãŒã‚ã‚‹

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from icecream import ic
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np

X,y=make_moons(n_samples=100,noise=0.25,random_state=3)
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=42)
forest=RandomForestClassifier(n_estimators=5,random_state=2)
forest.fit(X_train,y_train)

ic(forest.score(X_train,y_train))
ic(forest.score(X_test,y_test))

cmap=ListedColormap(('b','r'))

# è‰²åˆ†ã‘ç”¨ã®ãƒ¡ãƒƒã‚·ãƒ¥ãƒ‡ãƒ¼ã‚¿ã‚’ä½œã‚‹
x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1
x2_min, x2_max = X[:, 1].min()-1, X[:, 1].max()+1
x1_mesh, x2_mesh = np.meshgrid(np.arange(x1_min, x1_max, 0.01),np.arange(x2_min, x2_max,0.01))

## ãƒ¡ãƒƒã‚·ãƒ¥ãƒ‡ãƒ¼ã‚¿å…¨éƒ¨ã‚’å­¦ç¿’ãƒ¢ãƒ‡ãƒ«ã§åˆ†é¡ž
z = forest.predict(np.array([x1_mesh.ravel(), x2_mesh.ravel()]).T)
z = z.reshape(x1_mesh.shape)

## ãƒ¡ãƒƒã‚·ãƒ¥ãƒ‡ãƒ¼ã‚¿ã¨åˆ†é›¢ã‚¯ãƒ©ã‚¹ã‚’ä½¿ã£ã¦æ±ºå®šå¢ƒç•Œã‚’æã„ã¦ã„ã‚‹
plt.contourf(x1_mesh, x2_mesh, z, alpha=0.4, cmap=cmap)

plt.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap='jet',label='train_data',marker='o',alpha=0.5)
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap='jet',label='test_data',marker='x',alpha=0.5)
plt.legend()
plt.show()
```
