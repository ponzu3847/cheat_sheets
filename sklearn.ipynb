* ã“ã®é–¢æ•°ã¯å¼•æ•°ã«groupsã¨ã„ã†é…åˆ—ã‚’å–ã‚Šã€ã“ã®é…åˆ—ã«ã¯ç”»åƒã«å†™ã£ã¦ã„ã‚‹äººã‚’æŒ‡å®šã™ã‚‹ï¼ˆã‚¯ãƒ©ã‚¹ãƒ©ãƒ™ãƒ«ã¨ã¯é•ã†.ã“ã®å ´åˆã®ã‚¯ãƒ©ã‚¹ãƒ©ãƒ™ãƒ«ã¯ã€Œå¬‰ã—ã„ã€ãªã©ï¼‰
* åŒ»ç™‚ã‚¢ãƒ—ãƒªã‚±ãƒ¼ã‚·ãƒ§ãƒ³ãªã©ã§ã¯ã€åŒã˜æ‚£è€…ã‹ã‚‰å¾—ã‚‰ã‚ŒãŸè¤‡æ•°ã®ã‚µãƒ³ãƒ—ãƒ«ã‚’ç”¨ã„ã¦ã€æ–°ã—ã„æ‚£è€…ã«å¯¾ã™ã‚‹äºˆæ¸¬ã‚’è¡Œã†ã®ãŒç›®çš„
* ç™ºè©±èªè­˜ã§ã¯ã€åŒã˜è©±è€…ã«ã‚ˆã‚‹éŒ²éŸ³ãŒãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã«è¤‡æ•°å«ã¾ã‚Œã¦ã„ã‚‹å ´åˆã«ã€‚æ–°ã—ã„è©±è€…ã®ç™ºè©±ã‚’èªè­˜ã™ã‚‹ã“ã¨ãŒç›®çš„

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import GroupKFold

X,y=make_blobs(n_samples=20)
ic(y)
groups=np.random.randint(0,5,20)
ic(groups)
gkf=GroupKFold(n_splits=3)

for train_index, test_index in gkf.split(X,y,groups=groups):
    print("TRAIN:\n", train_index, "\nTEST:\n", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print("X_train.shape:{} X_test.shape:{} \ny_train.shape:{} y_test.shape:{}".format(X_train.shape,X_test.shape,y_train.shape,y_test.shape))
    print("np.bincount(y_train): {}\n".format(np.bincount(y_train)))
```

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV,KFold,train_test_split
import pandas as pd

boston=load_boston()
X_train,X_test,y_train,y_test=train_test_split(boston.data,boston.target,random_state=0)

kf=KFold(n_splits=5,shuffle=True,random_state=0)
param_grid={'alpha':[0.00001,0.0001,0.001,0.01,0.1,1,10,100]}
grid_search=GridSearchCV(Ridge(),param_grid,cv=kf)

grid_search.fit(X_train,y_train)

print("Test set score: {:.2f}".format(grid_search.score(X_test,y_test)))
print("Best parameters: {}".format(grid_search.best_params_))
print("Best cross_validation score: {:.2f}".format(grid_search.best_score_))
print("Best estimator:\n{}".format(grid_search.best_estimator_))

results=pd.DataFrame(grid_search.cv_results_)
display(results)

```

```python
from sklearn.datasets import load_boston
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

boston=load_boston()
X_train,X_test,y_train,y_test=train_test_split(boston.data,boston.target,random_state=0)
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

pca=PCA(n_components=1)
X_train_pca=pca.fit_transform(X_train_scaled)
X_test_pca=pca.transform(X_test_scaled)

svr=SVR()
svr.fit(X_train_pca,y_train)

ic(svr.score(X_train_pca,y_train))
ic(svr.score(X_test_pca,y_test))

line=np.linspace(-5,5,1000).reshape(-1,1)

plt.scatter(X_train_pca,y_train,alpha=.5,c='b',marker='o',label='train_data')
plt.scatter(X_test_pca,y_test,alpha=.5,c='b',marker='x',label='test_data')
plt.plot(line,svr.predict(line),c='r')
plt.xlabel('First principal component')
plt.ylabel('price')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from icecream import ic

cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)
tree=DecisionTreeClassifier(random_state=0,max_depth=4)
tree.fit(X_train,y_train)
ic(tree.score(X_train,y_train))
ic(tree.score(X_test,y_test))
display(cancer.feature_names)
display(tree.feature_importances_)  # ç‰¹å¾´é‡ã®é‡è¦åº¦ã‚’è¡¨ç¤º
```

* ã™ã¹ã¦ã®æ±ºå®šæœ¨ã«ã‚ˆã‚‹å›žå¸°ãƒ¢ãƒ‡ãƒ«ã¯å¤–æŒ¿ï¼ˆextrapolateï¼‰ãŒã§ããªã„

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import mglearn
import matplotlib.pyplot as plt
from icecream import ic
import numpy as np

X,y=mglearn.datasets.make_wave(n_samples=100)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

tree=DecisionTreeRegressor(random_state=0,max_depth=2).fit(X_train,y_train)
ic(tree.score(X_train,y_train))
ic(tree.score(X_test,y_test))


line=np.linspace(-3,3,1000).reshape(-1,1)
plt.scatter(X_train,y_train,label='train_data',marker='o',color='b')
plt.scatter(X_test,y_test,label='test_data',marker='x',color='r')
plt.plot(line,tree.predict(line),linestyle='-',color='g',label='predict')
plt.legend()
plt.show()
```

```python
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from icecream import ic

iris=load_iris()
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,stratify=iris.target,random_state=42)
tree=DecisionTreeClassifier(random_state=0,max_depth=4)
tree.fit(X_train,y_train)
ic(tree.score(X_train,y_train))
ic(tree.score(X_test,y_test))
display(iris.feature_names)
display(tree.feature_importances_)  # ç‰¹å¾´é‡ã®é‡è¦åº¦ã‚’è¡¨ç¤º

export_graphviz(tree,out_file='data/tree.dot',class_names=iris.target_names,feature_names=iris.feature_names,impurity=False,filled=True)
```

* å¤§ãã„ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã«å¯¾ã—ã¦ãƒ©ãƒ³ãƒ€ãƒ ãƒ•ã‚©ãƒ¬ã‚¹ãƒˆã‚’ä½œæˆã™ã‚‹ãŸã‚ã«ã¯æ™‚é–“ãŒã‹ã‹ã‚‹ãŒã€è¨ˆç®—æ©Ÿä¸Šã®CPUã‚³ã‚¢ï¼ˆn_jobsãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã§ã‚³ã‚¢æ•°ã‚’æŒ‡å®šï¼‰ã‚’è¤‡æ•°ä½¿ã£ã¦ä¸¦åˆ—åŒ–ã§ãã‚‹
* n_jobs=-1ã¨ã™ã‚‹ã¨è¨ˆç®—æ©Ÿä¸Šã®ã™ã¹ã¦ã®ã‚³ã‚¢ã‚’ä½¿ç”¨ã™ã‚‹
* åŸºæœ¬çš„ã«ãƒ©ãƒ³ãƒ€ãƒ æŽ¢ç´¢ã®ãŸã‚ã€random_stateã‚’å›ºå®šã—ãªã„ã¨æ§‹ç¯‰ã•ã‚Œã‚‹ãƒ¢ãƒ‡ãƒ«ãŒå¤§ããå¤‰ã‚ã‚‹ã“ã¨ã«æ³¨æ„
* ãƒ©ãƒ³ãƒ€ãƒ ãƒ•ã‚©ãƒ¬ã‚¹ãƒˆã¯ãƒ†ã‚­ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ãªã©ã®éžå¸¸ã«é«˜æ¬¡å…ƒã§ç–Žãªãƒ‡ãƒ¼ã‚¿ã«å¯¾ã—ã¦ã¯ã†ã¾ãæ©Ÿèƒ½ã—ãªã„å‚¾å‘ã«ã‚ã‚‹
* ç·šå½¢ãƒ¢ãƒ‡ãƒ«ã‚ˆã‚Šã‚‚å¤šãã®ãƒ¡ãƒ¢ãƒªã‚’æ¶ˆè²»ã—ã€è¨“ç·´ã‚‚äºˆæ¸¬ã‚‚é…ã„ãŸã‚ã€å®Ÿè¡Œæ™‚é–“ã‚„ãƒ¡ãƒ¢ãƒªãŒé‡è¦ãªã‚¢ãƒ—ãƒªã§ã¯ç·šå½¢ãƒ¢ãƒ‡ãƒ«ã‚’ä½¿ã£ãŸã»ã†ãŒè‰¯ã„
* n_estimatorsã¯å¤§ãã‘ã‚Œã°å¤§ãã„ã»ã©ã€ã‚ˆã‚Šå¤šãã®æ±ºå®šæœ¨ã®å¹³å‡ã‚’ã¨ã‚Šã€éŽå‰°é©åˆãŒä½Žæ¸›ã•ã‚Œã‚¢ãƒ³ã‚µãƒ³ãƒ–ãƒ«ãŒé ‘å¥ã«ãªã‚‹ã‚‹ãŸã‚ã€ã‚ˆã„ï¼ˆæ™‚é–“ã¨ãƒ¡ãƒ¢ãƒªã®ã‚ã‚‹é™ã‚Šå¤§ããã™ã‚‹ï¼‰
* max_featuresã¯å€‹ã€…ã®æ±ºå®šæœ¨ã®ä¹±æ•°æ€§ã‚’æ±ºå®šã™ã‚‹ã¨ã¨ã‚‚ã«ã€max_featuresãŒå°ã•ããªã‚‹ã¨éŽå‰°é©åˆãŒä½Žæ¸›ã™ã‚‹ï¼ˆä¸€èˆ¬ã«ã¯ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆå€¤ã‚’ä½¿ã†ã¨è‰¯ã„ï¼‰
* ã‚¯ãƒ©ã‚¹åˆ†é¡žã«ã¤ã„ã¦ã¯ã€max_features=sqrt(n_features)ã€å›žå¸°ã«ã¤ã„ã¦ã¯max_features=n_featuresã¨ãªã£ã¦ã„ã‚‹
* max_featuresã‚„max_leaf_nodesã‚’è¿½åŠ ã™ã‚‹ã¨æ€§èƒ½ãŒä¸ŠãŒã£ãŸã‚Šã€è¨“ç·´ã‚„äºˆæ¸¬ã«ã‹ã‹ã‚‹æ™‚é–“ãŒå¤§å¹…ã«ç¸®ã¾ã‚‹ã“ã¨ãŒã‚ã‚‹

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from icecream import ic
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np

X,y=make_moons(n_samples=100,noise=0.25,random_state=3)
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=42)
forest=RandomForestClassifier(n_estimators=5,random_state=2)
forest.fit(X_train,y_train)

ic(forest.score(X_train,y_train))
ic(forest.score(X_test,y_test))

cmap=ListedColormap(('b','r'))

# è‰²åˆ†ã‘ç”¨ã®ãƒ¡ãƒƒã‚·ãƒ¥ãƒ‡ãƒ¼ã‚¿ã‚’ä½œã‚‹
x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1
x2_min, x2_max = X[:, 1].min()-1, X[:, 1].max()+1
x1_mesh, x2_mesh = np.meshgrid(np.arange(x1_min, x1_max, 0.01),np.arange(x2_min, x2_max,0.01))

## ãƒ¡ãƒƒã‚·ãƒ¥ãƒ‡ãƒ¼ã‚¿å…¨éƒ¨ã‚’å­¦ç¿’ãƒ¢ãƒ‡ãƒ«ã§åˆ†é¡ž
z = forest.predict(np.array([x1_mesh.ravel(), x2_mesh.ravel()]).T)
z = z.reshape(x1_mesh.shape)

## ãƒ¡ãƒƒã‚·ãƒ¥ãƒ‡ãƒ¼ã‚¿ã¨åˆ†é›¢ã‚¯ãƒ©ã‚¹ã‚’ä½¿ã£ã¦æ±ºå®šå¢ƒç•Œã‚’æã„ã¦ã„ã‚‹
plt.contourf(x1_mesh, x2_mesh, z, alpha=0.4, cmap=cmap)

plt.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap='jet',label='train_data',marker='o',alpha=0.5)
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap='jet',label='test_data',marker='x',alpha=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import mglearn
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import Ridge

X,y=mglearn.datasets.make_wave(n_samples=40)
bins=np.linspace(-3,3,11)
display(bins)

which_bin=np.digitize(X,bins=bins)
display(X[:5])
display(which_bin[:5])

encoder=OneHotEncoder(sparse=False)
X_binned=encoder.fit_transform(which_bin)

display(X_binned[:5])

ridge=Ridge()
ridge.fit(X_binned,y)

line=np.linspace(-3,3,1000,endpoint=False).reshape(-1,1)
line_binned=encoder.transform(np.digitize(line,bins=bins))

plt.scatter(X,y)
plt.vlines(bins,ymin=-3,ymax=3,linestyle='--')
plt.plot(line,ridge.predict(line_binned))
plt.show()
```

#### .PolynomialFeatures
class sklearn.preprocessing.PolynomialFeatures(degree=2, *, interaction_only=False, include_bias=True, order='C')  
https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html?highlight=polynomial#sklearn.preprocessing.PolynomialFeatures  

* ç‰¹å¾´é‡è¡¨ç¾ã‚’ã‚ˆã‚Šè±Šã‹ã«ã™ã‚‹ä¸€ã¤ã®æ–¹æ³•ã¨ã—ã¦ã€ç‰¹ã«ç·šå½¢ãƒ¢ãƒ‡ãƒ«ã«æœ‰åŠ¹ãªã‚‚ã®ãŒã€å…ƒã®ãƒ‡ãƒ¼ã‚¿ã®äº¤äº’ä½œç”¨ç‰¹å¾´é‡ï¼ˆinteraction featureï¼‰ã¨å¤šé …å¼ç‰¹å¾´é‡ï¼ˆpolynomial featureï¼‰ã‚’åŠ ãˆã‚‹æ–¹æ³•


```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from icecream import ic

iris=load_iris()
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,random_state=0,stratify=iris.target)

scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

poly=PolynomialFeatures(degree=2)
X_train_poly=poly.fit_transform(X_train_scaled)
X_test_poly=poly.transform(X_test_scaled)

ic(X_train.shape)
ic(X_train_poly.shape)
print(poly.get_feature_names())

logreg=LogisticRegression()
logreg.fit(X_train_poly,y_train)

ic(logreg.score(X_train_poly,y_train))
ic(logreg.score(X_test_poly,y_test))
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.decomposition import NMF,PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt

iris=load_iris()
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,stratify=iris.target,random_state=0)

pca=PCA(n_components=3,whiten=True,random_state=0)
pca.fit(X_train)

nmf=NMF(n_components=2,random_state=0,max_iter=1000)
nmf.fit(X_train)

X_train_pca=pca.transform(X_train)
X_tset_pca=pca.transform(X_test)
X_train_nmf=nmf.transform(X_train)
X_test_nmf=nmf.transform(X_test)

svm_pca=LinearSVC(max_iter=10000).fit(X_train_pca,y_train)
svm_nmf=LinearSVC(max_iter=10000).fit(X_train_nmf,y_train)

ic(svm_pca.score(X_test_pca,y_test))
ic(svm_nmf.score(X_test_nmf,y_test))

fig,axes=plt.subplots(1,2,figsize=(16,6),facecolor='w')
for ax,model,model_name,data in zip(axes.ravel(),[svm_pca,svm_nmf],['PCA','NMF'],[[X_train_pca,X_test_pca],[X_train_nmf,X_test_nmf]]):
    ax.scatter(data[0][:,0],data[0][:,1],c=model.predict(data[0]),marker='o',cmap='viridis',label='train_data',alpha=0.5)
    ax.scatter(data[1][:,0],data[1][:,1],c=model.predict(data[1]),marker='x',cmap='viridis',label='test_data',alpha=0.5)
    ax.set_xlabel('Principal component 0')
    ax.set_ylabel('Principal component 1')
    ax.set_title(model_name)
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans

X,y=make_moons(n_samples=200,noise=0.05,random_state=0)

kmeans=KMeans(n_clusters=10,random_state=0)
kmeans.fit(X)
y_pred=kmeans.predict(X)

plt.scatter(X[:,0],X[:,1],c=y_pred,s=60,cmap='Paired')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=60,marker='^',c=range(kmeans.n_clusters),linewidths=2,cmap='Paired')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
display(y_pred)
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from icecream import ic
from collections import Counter
import matplotlib.pyplot as plt

iris=load_iris()
scaler=StandardScaler()
X_scaled=scaler.fit_transform(iris.data)
dbscan=DBSCAN(min_samples=15,eps=.7)
clusters=dbscan.fit_predict(X_scaled)

count=Counter(clusters)
ic(count)

feature_x=0
feature_y=2

for i in count.keys():
    plt.scatter(X_scaled[clusters==i,feature_x],X_scaled[clusters==i,feature_y],label='noise'if i==-1 else 'cluster_{}'.format(i))
plt.xlabel('Feature {}'.format(feature_x))
plt.ylabel('Feature {}'.format(feature_y))
plt.legend(loc='best')
plt.show()
```

æ­£è§£ãƒ‡ãƒ¼ã‚¿ã‚’ç”¨ã„ãŸã‚¯ãƒ©ã‚¹ã‚¿ãƒªãƒ³ã‚°ã®è©•ä¾¡
* ã‚¯ãƒ©ã‚¹ã‚¿ãƒªãƒ³ã‚°ã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ ã®å‡ºåŠ›ã‚’ã€æ­£è§£ãƒ‡ãƒ¼ã‚¿ã‚¯ãƒ©ã‚¹ã‚¿ãƒªãƒ³ã‚°ã¨æ¯”è¼ƒã—ã¦è©•ä¾¡ã™ã‚‹ãŸã‚ã«ç”¨ã„ã‚‰ã‚Œã‚‹æŒ‡æ¨™ãŒã„ãã¤ã‹ã‚ã‚Šã€æœ€ã‚‚é‡è¦ãªã®ã¯ã€èª¿æ•´ãƒ©ãƒ³ãƒ‰æŒ‡æ•°ï¼ˆadjusted rand index :ARIï¼‰ã¨æ­£è¦åŒ–ç›¸äº’æƒ…å ±é‡ï¼ˆnormalized mutual information :NMIï¼‰
* ã“ã‚Œã‚‰ã¯ã„ãšã‚Œã‚‚å®šé‡çš„ãªæŒ‡æ¨™ã§æœ€è‰¯ã®å ´åˆã«1ã‚’ã€é–¢ä¿‚ãªã„ã‚¯ãƒ©ã‚¹ã‚¿ãƒªãƒ³ã‚°ã®å ´åˆã«0ã‚’å–ã‚‹ï¼ˆãŸã ã—ARIã¯è² ã®å€¤ã«ãªã‚Šå¾—ã‚‹ï¼‰

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans,AgglomerativeClustering,DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score

X,y=make_blobs(n_samples=100,random_state=2)

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

fig,axes=plt.subplots(1,4,figsize=(30,6),facecolor='w')

axes[0].scatter(X_scaled[:,0],X_scaled[:,1],c=y)
axes[0].set_title('Correct label')

for ax,algorithm in zip(axes.ravel()[1:],[KMeans(n_clusters=3),AgglomerativeClustering(n_clusters=3),DBSCAN(min_samples=15,eps=.5)]):
    clusters=algorithm.fit_predict(X_scaled)
    score=adjusted_rand_score(y,clusters)
    ax.scatter(X_scaled[:,0],X_scaled[:,1],c=clusters)
    ax.set_title(algorithm.__class__.__name__+' (ARI: {:.2f})'.format(score))
plt.show()
```

* å€‹ã€…ã®ç‰¹å¾´é‡ã¨ã‚¿ãƒ¼ã‚²ãƒƒãƒˆã¨ã®é–“ã«çµ±è¨ˆçš„ã«é¡•è‘—ãªé–¢ä¿‚ãŒã‚ã‚‹ã‹ã©ã†ã‹ã‚’è¨ˆç®—ã—ã€æœ€ã‚‚é«˜ã„ç¢ºä¿¡åº¦ã§é–¢é€£ã—ã¦ã„ã‚‹ç‰¹å¾´é‡ãŒé¸æŠžã•ã‚Œã‚‹
* ã‚¯ãƒ©ã‚¹åˆ†é¡žã®å ´åˆã¯ã€åˆ†æ•£åˆ†æžï¼ˆanalysis of variance: ANOVAï¼‰ã¨ã—ã¦çŸ¥ã‚‰ã‚Œã‚‹æ‰‹æ³•
* ã“ã®æ–¹æ³•ã®ç‰¹æ€§ã¯å˜å¤‰é‡ã§ã‚ã‚‹ã“ã¨ã§ã‚ã‚Šã€ã¤ã¾ã‚Šå€‹ã€…ã®ç‰¹å¾´é‡ã‚’å€‹åˆ¥ã«è€ƒæ…®ã™ã‚‹ãŸã‚ã€ä»–ã®ç‰¹å¾´é‡ã¨çµ„ã¿åˆã‚ã•ã£ã¦æ„å‘³ã‚’æŒã¤ã‚ˆã†ãªç‰¹å¾´é‡ã¯æ¨ã¦ã‚‰ã‚Œã‚‹
* è¨ˆç®—ãŒé«˜é€Ÿã§ã€ãƒ¢ãƒ‡ãƒ«ã‚’æ§‹ç¯‰ã™ã‚‹å¿…è¦ãŒãªã„
* ä¸€æ–¹ã§ã€ç‰¹å¾´é‡é¸æŠžå¾Œã«åˆ©ç”¨ã™ã‚‹ãƒ¢ãƒ‡ãƒ«ã¨å®Œå…¨ã«ç‹¬ç«‹ã§ã‚ã‚‹

```python
from sklearn.datasets import load_boston
from sklearn.feature_selection import SelectPercentile
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

boston=load_boston()
X_train,X_test,y_train,y_test=train_test_split(boston.data,boston.target,random_state=0)

scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

lr=LinearRegression()
score=lr.fit(X_train_scaled,y_train).score(X_test_scaled,y_test)
ic(score)

select=SelectPercentile(percentile=10)
select.fit(X_train_scaled,y_train)
X_train_selected=select.transform(X_train_scaled)
X_test_selected=select.transform(X_test_scaled)
ic(X_train_scaled.shape,X_train_selected.shape,X_test_selected.shape)
mask=select.get_support()
selected_features=boston.feature_names[mask]
ic(selected_features)

lr=LinearRegression()
score=lr.fit(X_train_selected,y_train).score(X_test_selected,y_test)
ic(score)
```

```python
from sklearn.datasets import load_boston
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

boston=load_boston()
X_train,X_test,y_train,y_test=train_test_split(boston.data,boston.target,random_state=0)

scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

lr=LinearRegression()
score=lr.fit(X_train_scaled,y_train).score(X_test_scaled,y_test)
ic(score)

select=SelectFromModel(RandomForestRegressor(n_estimators=100,random_state=0),max_features=2,threshold='median')  # ã—ãã„å€¤ã¨ã—ã¦ä¸­å¤®å€¤ã‚’ç”¨ã„ã‚‹
select.fit(X_train_scaled,y_train)
X_train_selected=select.transform(X_train_scaled)
X_test_selected=select.transform(X_test_scaled)
ic(X_train_scaled.shape,X_train_selected.shape,X_test_selected.shape)
mask=select.get_support()
selected_features=boston.feature_names[mask]
ic(selected_features)

lr=LinearRegression()
score=lr.fit(X_train_selected,y_train).score(X_test_selected,y_test)
ic(score)
```

```python
from sklearn.datasets import load_boston
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

boston=load_boston()
X_train,X_test,y_train,y_test=train_test_split(boston.data,boston.target,random_state=0)

scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

lr=LinearRegression()
score=lr.fit(X_train_scaled,y_train).score(X_test_scaled,y_test)
ic(score)

select=RFE(RandomForestRegressor(n_estimators=100,random_state=0),n_features_to_select=2)   # ç‰¹å¾´é‡ã‚’ï¼’å€‹ã¾ã§çµžã‚‹
select.fit(X_train_scaled,y_train)
X_train_selected=select.transform(X_train_scaled)
X_test_selected=select.transform(X_test_scaled)
ic(X_train_scaled.shape,X_train_selected.shape,X_test_selected.shape)
mask=select.get_support()
selected_features=boston.feature_names[mask]
ic(selected_features)

lr=LinearRegression()
score=lr.fit(X_train_selected,y_train).score(X_test_selected,y_test)
ic(score)
```
