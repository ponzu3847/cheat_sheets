```
import copy
```

```
'''
copy
.deepcopy()

機能：
  多次元配列の深い層まで値渡しをする
使い方：
  copy.deepcopy(配列)
'''
x=[[1,2,3],[4,5,6]]
print('x=',x)
y=x
z=x.copy()
a=copy.copy(x)
b=copy.deepcopy(x)

x[1].append(10)
print('x[1].append(10)')
print('x=',x,'id(x[1])=',id(x[1]))
print('y=x ->',y,'id(y[1])=',id(y[1]))    #y[1]は参照渡し
print('x.copy() ->',z,'id(z[1])=',id(z[1]))   #z[1]は参照渡し
print('copy.copy(x) ->',a,'id(a[1])=',id(a[1]))   #a[1]は参照渡し
print('copy.deepcopy(x) ->',b,'id(b[1])=',id(b[1]))   #b[1]は値渡し
```
